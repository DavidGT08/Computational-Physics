# Pendulo simple




> Ecuaciones de movimiento del sistema


$\dfrac{d^2 \theta}{dt^2}+\dfrac{g}{l}\sin \theta=0$





```python
from pylab import *
import numpy as np
import numpy as np

'''
Pendulo simple

dTheta/dt^2 = -g/L sin(theta)

'''

#paramentros del sistema
g = 9.8 #m/s^2
L = 1 #m

#valores iniciales
theta_val , omega_val = np.meshgrid(np.arange(2, 10, 0.1), np.arange(-20, 20, 0.1))

#ecuaciones del sistema
theta_dot = omega_val
omega_dot = -g/L * np.sin(theta_val)


streamplot(theta_val, omega_val, theta_dot, omega_dot)
ylabel("Velocidad")
xlabel("Angulo (  )")
title('Espacio de fase [Pendulo simple]')
show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definicin de la ecuacin diferencial del pndulo simple
def pendulo_simple(t, y):
    # y[0] es el ngulo theta, y[1] es la velocidad angular dtheta/dt
    g = 9.81  # Aceleracin debido a la gravedad en m/s^2
    L = 1.0   # Longitud del pndulo en metros
    theta, theta_dot = y
    dydt = [theta_dot, -(g / L) * np.sin(theta)]
    return dydt

# Condiciones iniciales
theta0 = np.pi / 4.0  # ngulo inicial en radianes (45 grados)
theta_dot0 = 0.0      # Velocidad angular inicial

# Intervalo de tiempo para la integracin
t_span = (0.0, 10.0)  # Desde t=0 hasta t=10 segundos

# Resuelve la ecuacin diferencial usando solve_ivp
sol = solve_ivp(pendulo_simple, t_span, [theta0, theta_dot0], t_eval=np.linspace(0, 10, 1000))

# Extrae los resultados de la solucin
t = sol.t
theta = sol.y[0]
theta_dot = sol.y[1]

# Grafica el ngulo del pndulo en funcin del tiempo
plt.figure(figsize=(8, 4))
plt.plot(t, theta, label='$\\theta_0=\\pi/10$')
plt.plot(t, theta_dot, label='Velocidad Angular ($\\dot{\\theta}$)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Variable dinmica')
plt.title('Evolucin temporal de las variables [Pndulo simple]')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
# Extrae la velocidad angular de la solucin
theta_dot = sol.y[1]

# Grafica la velocidad angular en funcin del tiempo
plt.figure(figsize=(8, 4))

plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad Angular (radianes/s)')
plt.title('Velocidad Angular del Pndulo Simple')
plt.legend()
plt.grid()
plt.show()
```
