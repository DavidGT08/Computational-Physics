# Pendulo FÃ­sico

$\dfrac{d^2\theta}{dt^2}=-\dfrac{mgl}{I_s}\theta$


```python
from pylab import *
import numpy as np
import numpy as np

'''
Pendulo fisico

$\dfrac{d^2\theta}{dt^2}=-\dfrac{mgl}{I_s}\theta$


'''

#paramentros del sistema
l = 1 # longitud
g = 10 # $k$
m = 1 # mass
I = 1/3*m*l**2 # inertia momentum



#valores iniciales
theta_val , omega_val = np.meshgrid(np.arange(-1, 1, 0.1), np.arange(-10, 10, 0.1))

#ecuaciones del sistema
theta_dot = omega_val
omega_dot = -m*g*l/I*theta_val


streamplot(theta_val, omega_val, theta_dot, omega_dot)
ylabel("Velocidad")
xlabel("Angulo ( Î¸ )")
title('Espacio de fase [Pendulo fisico]')

show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# ParÃ¡metros del pÃ©ndulo fÃ­sico
m = 1.0         # Masa de la barra
l = 1.0         # Longitud de la barra
Is = (1/3) * m * l**2  # Momento de inercia respecto al eje de rotaciÃ³n
g = 9.81        # AceleraciÃ³n debido a la gravedad

# DefiniciÃ³n de la ecuaciÃ³n diferencial del pÃ©ndulo fÃ­sico
def pendulo_fisico(t, y):
    # y[0] es la posiciÃ³n angular theta, y[1] es la velocidad angular dtheta/dt
    theta, theta_dot = y
    dydt = [theta_dot, -(m * g * l / Is) * theta]
    return dydt

# Condiciones iniciales
theta0 = np.pi / 4.0  # Ãngulo inicial en radianes (45 grados)
theta_dot0 = 0.0      # Velocidad angular inicial

# Intervalo de tiempo para la integraciÃ³n
t_span = (0.0, 10.0)  # Desde t=0 hasta t=10 segundos

# Resuelve la ecuaciÃ³n diferencial usando solve_ivp
sol = solve_ivp(pendulo_fisico, t_span, [theta0, theta_dot0], t_eval=np.linspace(0, 10, 1000))

# Extrae los resultados de la soluciÃ³n
t = sol.t
theta = sol.y[0]
theta_dot = sol.y[1]

# Grafica la posiciÃ³n angular en funciÃ³n del tiempo
plt.figure(figsize=(10, 5))
# plt.subplot(2, 1, 1)
plt.plot(t, theta, label='Ãngulo ($\\theta$)')
plt.plot(t, theta_dot, label='Velocidad Angular ($\\dot{\\theta}$)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Variables')
plt.title('EvoluciÃ³n temporal de las variables [PÃ©ndulo fÃ­sico simple]')
plt.legend()
plt.grid()

# Grafica la velocidad angular en funciÃ³n del tiempo
# plt.subplot(2, 1, 2)
# plt.plot(t, theta_dot, label='Velocidad Angular (dtheta/dt)')
# plt.xlabel('Tiempo (s)')
# plt.ylabel('Velocidad Angular (radianes/s)')
# plt.legend()
# plt.grid()

plt.tight_layout()
plt.show()
```
