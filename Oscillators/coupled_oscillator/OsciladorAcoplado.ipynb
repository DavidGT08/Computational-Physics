# Oscilador Acoplado


> Ecuación para partícula 1

$m\dfrac{d^2x_1}{dt^2}=-(k+\kappa)x_1+κ x_2$

> Ecuación para partícula 2


$m\dfrac{d^2x_2}{dt^2}=-(k+\kappa)x_2+κ x_1$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del oscilador acoplado
omega1 = 1.0   # Frecuencia angular 1
omega2 = 2.0   # Frecuencia angular 2
alpha1 = 0.0   # Fase inicial 1
alpha2 = np.pi # Fase inicial 2

# Generar valores de tiempo t
t = np.linspace(0, 10, 1000)  # Intervalo de tiempo de 0 a 10 segundos con 1000 puntos

# Calcular las variables x1 y x2 en función de t
x1 = np.cos(omega1 * t + alpha1) + np.cos(omega2 * t + alpha2)
x2 = (np.sqrt(5) + 1) / 2 * np.cos(omega1 * t + alpha1) - (np.sqrt(5) + 1) / 2 * np.cos(omega2 * t + alpha2)

# Graficar x1 en función de t
plt.figure(figsize=(10, 5))
# plt.subplot(2, 1, 1)
plt.plot(t, x1, label='$x_1$')
plt.plot(t, x2, label='$x_2$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (x)')
plt.title('Evolucion temporal de la posicion [Oscilador acoplado]')
plt.legend()
plt.grid()

# Graficar x2 en función de t
# plt.subplot(2, 1, 2)
# plt.plot(t, x2, label='x2')
# plt.xlabel('Tiempo (s)')
# plt.ylabel('x2')
# plt.legend()
# plt.grid()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del oscilador acoplado
omega1 = 1.0   # Frecuencia angular 1
omega2 = 2.0   # Frecuencia angular 2
alpha1 = 0.0   # Fase inicial 1
alpha2 = np.pi # Fase inicial 2

# Generar valores de tiempo t
t = np.linspace(0, 10, 1000)  # Intervalo de tiempo de 0 a 10 segundos con 1000 puntos

# Calcular las variables x1 y x2 en función de t
x1 = np.cos(omega1 * t + alpha1) + np.cos(omega2 * t + alpha2)
x2 = (np.sqrt(5) + 1) / 2 * np.cos(omega1 * t + alpha1) - (np.sqrt(5) + 1) / 2 * np.cos(omega2 * t + alpha2)

# Calcular las velocidades v1 y v2 como derivadas temporales de x1 y x2
v1 = -omega1 * np.sin(omega1 * t + alpha1) - omega2 * np.sin(omega2 * t + alpha2)
v2 = -(np.sqrt(5) + 1) / 2 * omega1 * np.sin(omega1 * t + alpha1) + (np.sqrt(5) + 1) / 2 * omega2 * np.sin(omega2 * t + alpha2)

# Graficar v1 y v2 en función de t
plt.figure(figsize=(10, 5))
# plt.subplot(2, 1, 1)
plt.plot(t, v1, label='$v_1$')
plt.plot(t, v2, label='$v_2$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad ($\\dot{x}$)')
plt.title('Evolución Temporal de la Velocidad [Oscilador acoplado]')
plt.legend()
plt.grid()

# plt.subplot(2, 1, 2)
# plt.plot(t, v2, label='v2')
# plt.xlabel('Tiempo (s)')
# plt.ylabel('Velocidad v2')
# plt.legend()
# plt.grid()

plt.tight_layout()
plt.show()
```
