# Oscilador amortiguado

$\dfrac{d^2x}{dt^2}+2\gamma\dfrac{dx}{dt}+\omega_0^2 x=0$

donde


> $\omega_0^2=\dfrac{k}{m}$; $k$ es una constante del sistema


> $2\gamma=\dfrac{\lambda}{m}$; $\lambda$  es una constante proporcional a la fuerza de rozamiento







```python
from pylab import *
import numpy as np
import numpy as np

'''
Pendulo doble

\dfrac{d^2x}{dt^2}+2\gamma\dfrac{dx}{dt}+\omega_0^2 x=0

'''

#paramentros del sistema
l = 1 # $\lambda$
k = 1 # $k$
m = 1 # mass

omega2 = k/m
gamma  = 0.5

#valores iniciales
equis_val , vel_val = np.meshgrid(np.arange(-10, 10, 0.1), np.arange(-10, 10, 0.1))

#ecuaciones del sistema
equis_dot = vel_val
vel_dot = -2*gamma*equis_dot-omega2*equis_val


streamplot(equis_val, vel_val, equis_dot, vel_dot)
ylabel("Velocidad")
xlabel("Angulo (  )")
title('Espacio de fase [Oscilador amortiguado]')

show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parmetros del oscilador amortiguado
gamma = 0.1  # Coeficiente de amortiguamiento
omega0 = 1.0  # Frecuencia angular natural

# Definicin de la ecuacin diferencial del oscilador amortiguado
def oscilador_amortiguado(t, y):
    # y[0] es la posicin x, y[1] es la velocidad dx/dt
    x, x_dot = y
    dydt = [x_dot, -2 * gamma * x_dot - omega0**2 * x]
    return dydt

# Condiciones iniciales
x0 = 1.0  # Posicin inicial
x_dot0 = 0.0  # Velocidad inicial

# Intervalo de tiempo para la integracin
t_span = (0.0, 20.0)  # Desde t=0 hasta t=20 segundos

# Resuelve la ecuacin diferencial usando solve_ivp
sol = solve_ivp(oscilador_amortiguado, t_span, [x0, x_dot0], t_eval=np.linspace(0, 20, 1000))

# Extrae los resultados de la solucin
t = sol.t
x = sol.y[0]
x_dot = sol.y[1]

# Grafica la posicin en funcin del tiempo
plt.figure(figsize=(10, 5))
#plt.subplot(2, 1, 1)
plt.plot(t, x, label='Posicin (x)')
plt.plot(t, x_dot, label='Velocidad ($\\dot{x}$)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Variable')
plt.title('Evolucin temporal de las variables [Oscilador amortiguado]')
plt.legend()
plt.grid()

# Grafica la velocidad en funcin del tiempo
#plt.subplot(2, 1, 2)
# plt.plot(t, x_dot, label='Velocidad (dx/dt)')
# plt.xlabel('Tiempo (s)')
# plt.ylabel('Velocidad')
# plt.legend()
# plt.grid()

plt.tight_layout()
plt.show()
```
