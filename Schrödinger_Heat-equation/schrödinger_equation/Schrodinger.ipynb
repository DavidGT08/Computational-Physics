# Partícula en una caja

## 1 dimensión

```python
#Se resuelve la ecuación de Schrodinger por diferencias finitas
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Parámetros
L = 1.0
N = 1000
hbar = 1.0
m = 1.0
V0 = -10.0
a = L/N
x = np.linspace(0, L, N)

# Definir el potencial
V = np.zeros(N)
V[(N // 4):(3 * N // 4)] = V0

#Hamiltoniano
T = (-hbar**2 / (2 * m)) * (np.diag(-2.0 * np.ones(N)) + np.diag(np.ones(N - 1), 1) + np.diag(np.ones(N - 1), -1)) / (a**2)
H = T + np.diag(V)

#Resolver la ED
energias, funciones = np.linalg.eigh(H)

#Normalizar las funciones de onda
norms = np.sqrt(np.trapz(np.abs(funciones)**2, x, axis=0))
funciones_normalizadas = funciones / norms

#Graficar
plt.figure(figsize=(8, 6))
colour=iter(cm.plasma(np.linspace(0,3,12)))


for i in range(0,5,1):
    c= next(colour)
    plt.plot(x, funciones_normalizadas[:, i], label=f'E{str(i + 1)}', c=c, linewidth=2)


#plt.plot(x, V, 'k--' , label='Potencial')
plt.title('Partícula en la caja 1D')
plt.xlabel('x')
plt.ylabel('$\psi(x)$')
plt.legend()
plt.grid()
plt.show()
```

## 2 dimensiones

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

def Psi(L, x, y, n, A): #Funcion de Onda, L= longitud caja, A= cte de normalizacion
	Psi = (A*(np.sin((np.pi*n*x)/(L)))) * (A*(np.sin((np.pi*n*y)/(L))))
	return Psi

L = 10E-9

A = np.sqrt(2/L)

x = np.linspace(0,L,100)
y = np.linspace(0,L,100)
x, y = np.meshgrid(x,y)

n_array = np.linspace(1,3,3, dtype=int)

for n in n_array:
    i = int(n)
    z = Psi(L, x, y, n, A)
    fig = plt.figure()
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x, y, z, cmap="plasma")
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    ax.set_zlabel("$\Psi$")
    ax.set_title("Particula en la caja 2D, $n = %d $"%i)
    plt.show()


```

## 3 dimensiones

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

def Psi(L, x, y, z, n, A): #Funcion de Onda, L= longitud caja, A= cte de normalizacion
	Psi = (A*(np.sin((np.pi*n*x)/(L)))) * (A*(np.sin((np.pi*n*y)/(L)))) * (A*(np.sin((np.pi*n*z)/(L))))
	return Psi

L = 10E-9

A = (2/L)**(3/2)

x = np.linspace(0,L,100)
z = np.linspace(0,L,100)
y = np.linspace(0,L,100)
x1, y1, = np.meshgrid(x,y)
x2, z1, = np.meshgrid(x,z)
y2, z2, = np.meshgrid(y,z)
n_array = np.linspace(1,3,3, dtype=int)

for n in n_array:
    i = int(n)
    psi = Psi(L, x1, y2, z1, n, A)
    fig = plt.figure()
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x1, y1, psi, cmap="plasma")
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    ax.set_zlabel("$\Psi$")
    ax.set_title("Particula en la caja 3D, $n = %d $, $(x,y)$"%i)
    plt.show()
    ax = plt.axes(projection = "3d")
    ax.plot_surface(x2, z1, psi, cmap="plasma")
    plt.xlabel("$x$")
    plt.ylabel("$z$")
    ax.set_zlabel("$\Psi$")
    ax.set_title("Particula en la caja 3D, $n = %d $, $(x,z)$"%i)
    plt.show()
    ax = plt.axes(projection = "3d")
    ax.plot_surface(y2, z2, psi, cmap="plasma")
    plt.xlabel("$y$")
    plt.ylabel("$z$")
    ax.set_zlabel("$\Psi$")
    ax.set_title("Particula en la caja 3D, $n = %d $, $(y,z)$"%i)
    plt.show()



```
